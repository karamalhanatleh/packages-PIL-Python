# How to open an image with PIL

```python
# import
from PIL import Image
import matplotlib.pyplot as plt
```

### How to create new image


```python
new = Image.new(mode="RGBA", size=(1920,1080))
new.show()
```

```python
new= Image.new("RGBA" , (1920 , 1080))
new.show()
```

### How to create new image with color name

```python
new = Image.new(mode="RGBA", size=(1920,1080), color="pink")
new.show()
```

### How to create new image with RGB color

```python
											#          r  g  b   alpha
new=  Image.new(mode ="RGBA"  ,size=(520,520) , color=(10,10,255,155))
new
```

### How to create new image with hex color

```python
new=Image.new(mode="RGB" , size=(520,520) , color="#00004c")
new
```
